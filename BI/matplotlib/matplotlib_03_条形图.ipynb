# 条形图
- plt.bar(x, height, width=0.8)  # 绘制纵向条形图

## 案例3
假设你知道了列表a中电影分别在2017-09-14(b_14), 2017-09-15(b_15), 2017-09-16(b_16)三天的票房,为了展示列表中电影本身的票房以及同其他电影的数据对比情况,应该如何更加直观的呈现该数据?

a = ["猩球崛起3：终极之战","敦刻尔克","蜘蛛侠：英雄归来","战狼2"]  
b_16 = [15746,312,4497,319]  
b_15 = [12357,156,2045,168]  
b_14 = [2358,399,2358,362]


```python
from matplotlib import pyplot as plt
from matplotlib import font_manager


# 设置中文字体
my_font = font_manager.FontProperties(fname="/usr/share/fonts/truetype/arphic/uming.ttc")

a = ["猩球崛起3：终极之战","敦刻尔克","蜘蛛侠：英雄归来","战狼2"]
b_16 = [15746,312,4497,319]
b_15 = [12357,156,2045,168]
b_14 = [2358,399,2358,362]


bar_width = 0.2

x_14 = list(range(len(a)))
x_15 = [i+bar_width for i in x_14]
x_16 = [i+bar_width*2 for i in x_14]

# 设置图形大小
plt.figure(figsize=(8,4), dpi=100)

plt.bar(x_14, b_14, width=bar_width, label="9月14号")
plt.bar(x_15, b_15, width=bar_width, label="9月15号")
plt.bar(x_16, b_16, width=bar_width, label="9月16号")

# 设置刻度
plt.xticks(x_15, a, fontproperties=my_font)

# 设置图例
plt.legend(prop=my_font)

# 设置图形描述
plt.title("9月14-16日票房统计图", fontproperties=my_font)
plt.xlabel("电影名称", fontproperties=my_font)
plt.ylabel("票房数（万元）", fontproperties=my_font)

plt.show()
```
